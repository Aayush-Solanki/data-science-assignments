```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway, chi2_contingency


df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignments\hypothysis testing\LabTAT.csv")


print("Dataset Overview:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())


plt.figure(figsize=(10, 6))
for column in df.columns:
    sns.histplot(data=df, x=column, kde=True, label=column)
plt.title('Distribution of Turnaround Time (TAT) for Each Laboratory')
plt.xlabel('Turnaround Time (TAT)')
plt.ylabel('Frequency')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Turnaround Time (TAT) for Each Laboratory')
plt.ylabel('Turnaround Time (TAT)')
plt.show()


f_statistic, p_value_anova = f_oneway(df['Laboratory 1'], df['Laboratory 2'], df['Laboratory 3'], df['Laboratory 4'])
print(f'One-Way ANOVA F-statistic: {f_statistic}')
print(f'One-Way ANOVA p-value: {p_value_anova}')


if 'Existing_Categorical_Variable' in df.columns:
    contingency_table = pd.crosstab(df['Existing_Categorical_Variable'], df['Laboratory 1'])
    chi2_statistic, p_value_chi2, _, _ = chi2_contingency(contingency_table)
    print(f'Chi-Square Test p-value: {p_value_chi2}')
else:
    print('Existing_Categorical_Variable not found in the dataset.')
```
