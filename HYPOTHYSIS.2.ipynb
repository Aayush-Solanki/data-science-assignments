```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway


df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\hypothysis testing\LabTAT.csv")


print("Dataset Overview:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())


plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Laboratory 1', kde=True, color='blue', label='Laboratory 1')
sns.histplot(data=df, x='Laboratory 2', kde=True, color='red', label='Laboratory 2')
sns.histplot(data=df, x='Laboratory 3', kde=True, color='green', label='Laboratory 3')
sns.histplot(data=df, x='Laboratory 4', kde=True, color='orange', label='Laboratory 4')
plt.title('Distribution of Turnaround Time (TAT) for Each Laboratory')
plt.xlabel('Turnaround Time (TAT)')
plt.ylabel('Frequency')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Turnaround Time (TAT) for Each Laboratory')
plt.ylabel('Turnaround Time (TAT)')
plt.show()


f_statistic, p_value = f_oneway(df['Laboratory 1'], df['Laboratory 2'], df['Laboratory 3'], df['Laboratory 4'])
print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')

if p_value < 0.05:
    print('Reject the null hypothesis. There is a significant difference in average TAT among the laboratories.')
else:
    print('Fail to reject the null hypothesis. There is no significant difference in average TAT among the laboratories.')
```
