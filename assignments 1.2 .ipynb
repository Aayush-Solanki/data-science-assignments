```python
import matplotlib.pyplot as plt


boxplot1_data = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
boxplot2_data = [30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]


fig, ax = plt.subplots()


ax.boxplot([boxplot1_data, boxplot2_data], labels=['Boxplot 1', 'Boxplot 2'])


ax.set_xlabel('Boxplots')
ax.set_ylabel('Values')
ax.set_title('Comparison of Boxplot 1 and Boxplot 2')


plt.show()
```
