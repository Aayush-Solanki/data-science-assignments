```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


np.random.seed(0)
chick_weights = np.random.normal(loc=200, scale=50, size=100) 


plt.figure(figsize=(8, 6))
sns.boxplot(x=chick_weights)
plt.title('Boxplot of Chick Weights')
plt.xlabel('Chick Weights (grams)')
plt.show()


plt.figure(figsize=(8, 6))
sns.histplot(chick_weights, bins=20, kde=True)  
plt.title('Histogram of Chick Weights')
plt.xlabel('Chick Weights (grams)')
plt.ylabel('Frequency')
plt.show()
```
