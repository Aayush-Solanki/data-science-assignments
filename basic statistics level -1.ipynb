```python
# answer no.7
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignment-01\Q7.csv")

points = df['Points'].values
score = df['Score'].values
weight = df['Weigh'].values

def calculate_statistics(data):
    mean = np.mean(data)
    median = np.median(data)
    mode = stats.mode(data).mode
    variance = np.var(data)
    std_dev = np.std(data)
    data_range = np.max(data) - np.min(data)

    return mean, median, mode, variance, std_dev, data_range

points_stats = calculate_statistics(points)
score_stats = calculate_statistics(score)
weight_stats = calculate_statistics(weight)

print("Points:")
print("Mean:", points_stats[0])
print("Median:", points_stats[1])
print("Mode:", points_stats[2])
print("Variance:", points_stats[3])
print("Standard Deviation:", points_stats[4])
print("Range:", points_stats[5])

print("\nScore:")
print("Mean:", score_stats[0])
print("Median:", score_stats[1])
print("Mode:", score_stats[2])
print("Variance:", score_stats[3])
print("Standard Deviation:", score_stats[4])
print("Range:", score_stats[5])

print("\nWeight:")
print("Mean:", weight_stats[0])
print("Median:", weight_stats[1])
print("Mode:", weight_stats[2])
print("Variance:", weight_stats[3])
print("Standard Deviation:", weight_stats[4])
print("Range:", weight_stats[5])

```

```python
#answer no. 9
import pandas as pd
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

data_a = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignment-01\Q9_a.csv")
data_b = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignment-01\Q9_b.csv")

def calculate_skewness_kurtosis(data, column_name):
    skewness = skew(data[column_name])
    kurt = kurtosis(data[column_name])
    return skewness, kurt

speed_skewness, speed_kurtosis = calculate_skewness_kurtosis(data_a, 'speed')
distance_skewness, distance_kurtosis = calculate_skewness_kurtosis(data_a, 'dist')

sp_skewness, sp_kurtosis = calculate_skewness_kurtosis(data_b, 'SP')
wt_skewness, wt_kurtosis = calculate_skewness_kurtosis(data_b, 'WT')


print("Q9_a.csv - Cars Speed and Distance:")
print("Speed - Skewness:", speed_skewness, "Kurtosis:", speed_kurtosis)
print("Distance - Skewness:", distance_skewness, "Kurtosis:", distance_kurtosis)

print("\nQ9_b.csv - SP and Weight:")
print("SP - Skewness:", sp_skewness, "Kurtosis:", sp_kurtosis)
print("WT - Skewness:", wt_skewness, "Kurtosis:", wt_kurtosis)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data_a['speed'], kde=True)
plt.title('Cars Speed Histogram')

plt.subplot(1, 2, 2)
sns.histplot(data_a['dist'], kde=True)
plt.title('Cars Distance Histogram')

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data_b['SP'], kde=True)
plt.title('SP Histogram')

plt.subplot(1, 2, 2)
sns.histplot(data_b['WT'], kde=True)
plt.title('Weight Histogram')

plt.tight_layout()
plt.show()
```

```python
#answer no. 21(a)
import pandas as pd
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignment-01\Cars.csv")
mpg = df['MPG']

stat, p_value = shapiro(mpg)


print("Shapiro-Wilk Test:")
print(f"Statistic: {stat}, p-value: {p_value}")

plt.figure(figsize=(8, 5))
sns.histplot(mpg, kde=True, color='blue')
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
# answer no. 21(b)
import pandas as pd
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignment-01\wc-at.csv")

at = df['AT']
waist = df['Waist']


def check_normality(data, label):
    
    stat, p_value = shapiro(data)

    
    print(f"\nShapiro-Wilk Test for {label}:")
    print(f"Statistic: {stat}, p-value: {p_value}")

    
    plt.figure(figsize=(10, 6))
    sns.histplot(data, kde=True, color='blue')
    plt.title(f'Distribution of {label}')
    plt.xlabel(label)
    plt.ylabel('Frequency')
    plt.show()

check_normality(at, 'Adipose Tissue (AT)')

check_normality(waist, 'Waist Circumference (Waist)')
```
