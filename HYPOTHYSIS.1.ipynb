```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene, f_oneway


df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignments\hypothysis testing\Cutlets.csv)
                 
print("Dataset Overview:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())


plt.figure(figsize=(10, 6))
sns.histplot(data=df, kde=True)
plt.title('Distribution of Cutlet Diameters for Each Unit')
plt.xlabel('Cutlet Diameter')
plt.ylabel('Frequency')
plt.legend(df.columns)
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient='v')
plt.title('Boxplot of Cutlet Diameters for Each Unit')
plt.ylabel('Cutlet Diameter')
plt.show()


for col in df.columns:
    if col != 'Quality':
        _, p_value = shapiro(df[col])
        print(f'Shapiro-Wilk Test p-value for {col}: {p_value}')


_, p_value_levene = levene(df['Unit A'], df['Unit B'])
print(f'Levene\'s Test p-value: {p_value_levene}')

#
_, p_value_anova = f_oneway(df['Unit A'], df['Unit B'])
print(f'One-Way ANOVA p-value: {p_value_anova}')
```
