```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\association rules\my_movies.csv")
df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\association rules\book.csv")
def generate_and_visualize_rules(df, support_thresholds, confidence_thresholds, min_lengths):
    for min_length in min_lengths:
        for support_threshold in support_thresholds:
            for confidence_threshold in confidence_thresholds:
                
                frequent_itemsets = apriori(df, min_support=support_threshold, use_colnames=True)

                
                rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence_threshold)

                
            
                plt.figure(figsize=(8, 6))
                plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
                plt.xlabel('Support')
                plt.ylabel('Confidence')
                plt.title(f'Association Rules (min_length={min_length}, support={support_threshold}, confidence={confidence_threshold})')
                plt.show()

support_thresholds = [0.1, 0.2, 0.3]
confidence_thresholds = [0.5, 0.6, 0.7]
min_lengths = [2, 3]  
generate_and_visualize_rules(df, support_thresholds, confidence_thresholds, min_lengths)
```
