```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene, ttest_ind


df = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\hypothysis testing\Cutlets.csv")

print("Dataset Overview:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())


plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Unit A', kde=True, color='blue', label='Unit A')
sns.histplot(data=df, x='Unit B', kde=True, color='red', label='Unit B')
plt.title('Distribution of Cutlet Diameters for Each Unit')
plt.xlabel('Cutlet Diameter')
plt.ylabel('Frequency')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Unit A', 'Unit B']], orient='v')
plt.title('Boxplot of Cutlet Diameters for Each Unit')
plt.ylabel('Cutlet Diameter')
plt.show()


_, p_value_unit1 = shapiro(df['Unit A'])
_, p_value_unit2 = shapiro(df['Unit B'])
print(f'Shapiro-Wilk Test p-value for Unit A: {p_value_unit1}')
print(f'Shapiro-Wilk Test p-value for Unit B: {p_value_unit2}')


_, p_value_levene = levene(df['Unit A'], df['Unit B'])
print(f'Levene\'s Test p-value: {p_value_levene}')


t_statistic, p_value_ttest = ttest_ind(df['Unit A'], df['Unit B'])
print(f'Two-Sample T-Test p-value: {p_value_ttest}')
```
