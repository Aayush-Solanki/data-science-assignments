```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv(r"C:\Users\Owner\OneDrive\Desktop\data science assingment\assignments\RANDOM FOREST\Fraud_check.csv")


print("First few rows of the dataset:")
print(data.head())


print("\nSummary statistics of numerical features:")
print(data.describe())


print("\nMissing values:")
print(data.isnull().sum())


plt.figure(figsize=(8, 6))
sns.countplot(data['Taxable.Income'])
plt.title('Distribution of Taxable Income')
plt.xlabel('Taxable Income')
plt.ylabel('Count')
plt.show()


numeric_data = data.select_dtypes(include=['int64', 'float64'])


plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()


plt.figure(figsize=(12, 10))
for i, col in enumerate(numeric_data.columns):
    plt.subplot(3, 3, i+1)
    sns.histplot(numeric_data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 10))
for i, col in enumerate(data.select_dtypes(include=['object']).columns):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x=col, y='Taxable.Income', data=data)
    plt.title(f'Boxplot of Taxable Income by {col}')
    plt.xlabel(col)
    plt.ylabel('Taxable Income')
plt.tight_layout()
plt.show()


sns.pairplot(data=numeric_data)
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()


plt.figure(figsize=(12, 10))
for i, col in enumerate(data.select_dtypes(include=['object']).columns):
    plt.subplot(3, 3, i+1)
    sns.countplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 10))
for i, col in enumerate(numeric_data.columns):
    plt.subplot(3, 3, i+1)
    sns.boxplot(data[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```
